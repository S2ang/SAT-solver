```python
from collections import defaultdict
from itertools import combinations

class Front:
    
    def __init__(self, nodes, treasures, steps, paths):
        
        self.nodes = nodes
        self.treasures = treasures # dict key = node, value = treasure
        self.steps = steps
        self.paths = paths # dict key = node, value = neighbors
        
        self.at = defaultdict(list) # dict key = time I, value = nodes 
        self.has = defaultdict(list) # dict key = time I, value = treasures
        self.prop = {} # dict key = (node or treasure, time I), value = idx
        
        idx = 1
        for i in range(0, steps+1):
            for j in nodes:
                self.prop[(j, i)] = idx # (node, time I)
                self.at[i].append(j)
                idx += 1
        for i in range(0, steps+1):
            for j in treasures:
                self.prop[(treasures[j], i)] = idx # (treasures, time I)
                self.has[i].append(treasures[j])
                idx += 1      
    
    def getProp(self):
        return self.prop
    
    def category1(self):
        
        cate1 = ""
        
        for i in self.at:
            # all possible pairs in At()
            arr = [(x, i) for x in self.at[i]]
            res = list(combinations(arr, 2))
            for j in res:
                cate1 += f"-{self.prop[j[0]]} -{self.prop[j[1]]}" 
                cate1 += "\n"
        
        return cate1
                
    def category2(self):
        
        cate2 = ""
        
        for i in range(0, self.steps): 
            for a in self.at[i]:
                cate2 += f"-{self.prop[(a, i)]} "
                for n in self.paths[a]:
                    cate2 += f"{self.prop[(n, i+1)]} "
                cate2 += "\n"
        
        return cate2
            
    def category3(self):
        
        cate3 = ""
        
        for i in range(0, self.steps+1):
            for node in self.at[i]:
                if node in self.treasures:
                    cate3 += f"-{self.prop[(node, i)]} {self.prop[(self.treasures[node], i)]}"
                    cate3 += "\n"
        
        return cate3
            
    
    def category4(self):
        
        cate4 = ""
        
        for i in range(0, self.steps):
            for treasure in self.has[i]:
                cate4 += f"-{self.prop[(treasure, i)]} {self.prop[(treasure, i+1)]}"
                cate4 += "\n"
        
        return cate4
        
    def category5(self):
        
        cate5 = ""
        
        for i in range(0, self.steps):
            for treasure in self.has[i]:
                cate5 += f"{self.prop[(treasure, i)]} -{self.prop[(treasure, i+1)]} "
                for node in self.treasures:
                    if self.treasures[node] == treasure:
                        cate5 += f"{self.prop[(node, i+1)]} "
                cate5 += "\n"
        
        return cate5
    
    def category6(self):
        
        cate6 = f"{self.prop[('START', 0)]}"
        cate6 += "\n"
                            
        return cate6
    
    def category7(self):
                                  
        cate7 = ""
        
        for treasure in self.has[0]:
            cate7 += f"-{self.prop[(treasure, 0)]}"
            cate7 += "\n"
                                  
        return cate7
    
    def category8(self):
        
        cate8 = ""
        
        for treasure in self.has[self.steps]:
            cate8 += f"{self.prop[(treasure, self.steps)]}"
            cate8 += "\n"
        
        return cate8
                                  
    def formatting(self):
        
        cate1 = self.category1()
        cate2 = self.category2()
        cate3 = self.category3()
        cate4 = self.category4()
        cate5 = self.category5()
        cate6 = self.category6()
        cate7 = self.category7()
        cate8 = self.category8()
        
        with open('FrontEndOutput.txt', 'w') as f:
            
            f.writelines([cate1, cate2, cate3, cate4, cate5, cate6, cate7, cate8])
            f.write('0\n')
            
            for i in self.prop:
                
                if i[0] in self.nodes:
                    f.write(f"{self.prop[i]} At({i[0]},{i[1]})")
                    f.write('\n')
                else:
                    f.write(f"{self.prop[i]} Has({i[0][0]},{i[1]})")
                    f.write('\n')
        
```
