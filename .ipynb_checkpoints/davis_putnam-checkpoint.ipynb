```python
from collections import defaultdict

class Solver:
    
    def __init__(self):
        
        self.CS = []
        self.B = {}
        self.atoms = defaultdict(list)
        self.prop = {}
        self.keys = []
    
    def getInput(self, txt):
        
        with open(txt, 'r') as f:
            data = f.readlines()
            
            idx = data.index('0\n')
            cnf = []
            for i in data[:idx]:
                
                d = i.strip().split()
                new_d = [int(i) for i in d]
                
                cnf.append(new_d)
                
            for i in cnf:
                self.CS.append(i)
        
            for item in cnf:
                for i in item:
                    self.atoms[abs(i)].append(i)

            for i in self.atoms:
                self.atoms[i] = list(set(self.atoms[i]))
                
            print(self.atoms)
            
            self.keys = [i for i in self.atoms]
                        
            idx_2 = 1
            for i in data[idx+1:]:
                p = i.split()
                self.prop[idx_2] = p[1]
                idx_2 += 1      
    
    def getB(self):
        return self.B
    
    def getCS(self):
        return self.CS
    
    def getProp(self):
        return self.prop
    
    def updateAtom(self):
        self.atoms = defaultdict(list)
        
        for item in self.CS:
            for i in item:
                self.atoms[abs(i)].append(i)
        
        for i in self.atoms:
            self.atoms[i] = list(set(self.atoms[i]))
    
    def propagate(self, CS, B, A, V):
        
        B[A] = V
        
        if A in self.keys:
            self.keys.remove(A)
        
        temp = [i for i in CS]
        
        for c in temp:
            if A*V in c:
                CS.remove(c)
            elif (-1)*A*V in c:
                i = CS.index(c)
                CS[i].remove((-1)*A*V)
                
        self.updateAtom()
        
        return CS, B
    
    def easyCaseIn(self, CS, B, atoms):
        
        for c in CS:
            if len(c) == 1:
                return True
        
        for i in atoms:
            if len(atoms[i]) == 1:
                return True
            
        return False
            
    def atomOf(self, L):
        return abs(L)
    
    def valueOf(self, L):
        return L//abs(L)
        
    def easyCase(self, CS, B, atoms):
        
        for c in CS:
            
            if len(c) == 1:
                return self.propagate(CS, B, self.atomOf(c[0]), self.valueOf(c[0]))
        
        for i in atoms:
            
            if len(atoms[i]) == 1:
                return self.propagate(CS, B, i, self.valueOf(atoms[i][0]))
            
    
    def dpll(self, CS, B):
        if not CS:
            return B
        if [] in CS:
            return False
        if self.easyCaseIn(CS, B, self.atoms):
            self.CS, self.B = self.easyCase(CS, B, self.atoms)
        else:
            P = self.keys[0]

            CS, B = self.propagate(CS, B, P, 1);
            answer = self.dpll(CS, B)
            if (answer != False):
                return answer
            CS, B = self.propagate(CS, B, P, -1);
        
        return self.dpll(CS, B)
    
    def formatting(self):
        
        print(self.CS)
        
        with open('DPOutput.txt', 'w') as f:
            
            if self.B:
                for i in range(1, len(self.B)+1): 
                    if self.B[i] == 1:
                        f.write(f"{i} T")
                        f.write('\n')
                    else:
                        f.write(f"{i} F")
                        f.write('\n')

                f.write('0')
                f.write('\n')

                for i in self.prop:
                    f.write(f"{i} {self.prop[i]}")
                    f.write('\n')
            else:
                f.write("NO SOLUTION")
                
       
    
```
